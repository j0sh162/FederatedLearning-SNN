```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../../acc_hist.csv")
df.head()
```

```python
fig = plt.figure(figsize=(8, 5))
df["epoch_batch"] = df["epoch"].astype(str) + "-" + df["batch"].astype(str)
sns.lineplot(data=df, x="batch", y="Acc", hue="Set")
plt.title("Train/ Test Set Accuracy with 10 test batches")
plt.xlabel("Batch")
# plt.xticks(rotation=45, size=7)
plt.ylabel("Accuracy")
plt.show()
```
